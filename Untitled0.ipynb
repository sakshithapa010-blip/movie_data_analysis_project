```python
import pandas as pd

# 1. Basic Math: Profit and ROI
# We use .fillna(0) just in case some rows have empty budget/revenue
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0)
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce').fillna(0)

df['Profit'] = df['revenue'] - df['budget']
# We use a small trick to avoid dividing by zero if budget is 0
df['ROI'] = df.apply(lambda x: x['Profit'] / x['budget'] if x['budget'] > 0 else 0, axis=1)

# --- QUESTION 1: Highest Profit Movie ---
# Sorting by Profit
top_movie = df.sort_values(by='Profit', ascending=False).iloc[0]

print("--- 1. HIGHEST PROFIT MOVIE ---")
print(f"Movie Title: {top_movie['title']}")
print(f"Profit: ${top_movie['Profit']:,.2f}")
# Note: Since 'crew' and 'cast' are complex, we print the raw text for now
print(f"Cast Data: {top_movie['cast'][:200]}...")
print("-" * 30)

# --- QUESTION 2: ROI by Language ---
# 'original_language' is the column name in your data
lang_roi = df.groupby('original_language')['ROI'].mean().sort_values(ascending=False)

print("\n--- 2. HIGHEST AVERAGE ROI LANGUAGE ---")
print(lang_roi.head(1))
print("-" * 30)

# --- QUESTION 3: Unique Genres ---
# Your genres look like [{'id': 18, 'name': 'Drama'}, ...]
# This code extracts just the names
import ast

def get_genre_names(genre_str):
    try:
        # Converts string representation of list to actual list
        genres = ast.literal_eval(genre_str)
        return [g['name'] for g in genres]
    except:
        return []

# Apply the function and get unique values
all_genres = df['genres'].apply(get_genre_names).explode().unique()

print("\n--- 3. UNIQUE GENRES ---")
print(list(all_genres))
```

```python
import ast

# --- Helper Functions to extract names from the messy 'crew' and 'cast' columns ---
def get_director(crew_str):
    try:
        crew_list = ast.literal_eval(crew_str)
        return [person['name'] for person in crew_list if person['job'] == 'Director'][0]
    except:
        return "Unknown"

def get_producers(crew_str):
    try:
        crew_list = ast.literal_eval(crew_str)
        return [person['name'] for person in crew_list if person['job'] == 'Producer']
    except:
        return []

def get_actors(cast_str):
    try:
        cast_list = ast.literal_eval(cast_str)
        return [person['name'] for person in cast_list]
    except:
        return []

# Apply these functions to your data
df['director'] = df['crew'].apply(get_director)
df['producers_list'] = df['crew'].apply(get_producers)
df['actors_list'] = df['cast'].apply(get_actors)

# --- QUESTION 4: Top 3 Producers by ROI ---
# We explode the list because one movie can have multiple producers
producers_df = df.explode('producers_list')
top_producers = producers_df.groupby('producers_list')['ROI'].mean().sort_values(ascending=False).head(3)

print("--- 4. TOP 3 PRODUCERS (BY AVG ROI) ---")
print(top_producers)
print("-" * 30)

# --- QUESTION 5: Most Frequent Actor & Their Profits ---
# Find which actor appears most across all movies
all_actors = df.explode('actors_list')
top_actor = all_actors['actors_list'].value_counts().idxmax()

# Filter data for just this actor
actor_movies = all_actors[all_actors['actors_list'] == top_actor]
avg_profit = actor_movies['Profit'].mean()

print(f"\n--- 5. MOST FREQUENT ACTOR: {top_actor} ---")
print(f"Number of Movies: {len(actor_movies)}")
print(f"Average Profit per Movie: ${avg_profit:,.2f}")
print("-" * 30)

# --- QUESTION 6: Top 3 Directors' Favorite Actors ---
top_3_directors = df['director'].value_counts().head(3).index

print("\n--- 6. TOP 3 DIRECTORS & THEIR PREFERRED ACTORS ---")
for director in top_3_directors:
    director_films = df[df['director'] == director]
    fav_actor = director_films.explode('actors_list')['actors_list'].value_counts().idxmax()
    print(f"Director: {director} | Prefers Actor: {fav_actor}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set(style="whitegrid")

# 1. Prepare Data: Let's look at the Top 10 Languages by ROI
top_10_languages = df.groupby('original_language')['ROI'].mean().sort_values(ascending=False).head(10)

# 2. Create the Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_languages.index, y=top_10_languages.values, palette="viridis")

# 3. Add Labels
plt.title('Top 10 Languages by Average Return on Investment (ROI)', fontsize=16)
plt.xlabel('Language Code', fontsize=12)
plt.ylabel('Average ROI', fontsize=12)

plt.show()
```

```python
import pandas as pd
import ast

# 1. Clean data and define key columns again to be safe
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0)
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce').fillna(0)
df['Profit'] = df['revenue'] - df['budget']
df['ROI'] = df.apply(lambda x: x['Profit'] / x['budget'] if x['budget'] > 0 else 0, axis=1)

# 2. Extract Names (Director, Producers, Actors)
def get_job_name(crew_str, job_title):
    try:
        crew_list = ast.literal_eval(crew_str)
        names = [person['name'] for person in crew_list if person['job'] == job_title]
        return names[0] if job_title == 'Director' and names else names
    except:
        return "Unknown" if job_title == 'Director' else []

def get_actors(cast_str):
    try:
        return [person['name'] for person in ast.literal_eval(cast_str)]
    except:
        return []

df['director'] = df['crew'].apply(lambda x: get_job_name(x, 'Director'))
df['producers_list'] = df['crew'].apply(lambda x: get_job_name(x, 'Producer'))
df['actors_list'] = df['cast'].apply(get_actors)

# 3. Calculate the variables for the report
top_movie = df.sort_values(by='Profit', ascending=False).iloc[0]
language_roi = df.groupby('original_language')['ROI'].mean().sort_values(ascending=False)
all_genres = df['genres'].apply(get_actors).explode().unique() # Using same logic for genres

# Top Producers
producers_df = df.explode('producers_list')
top_producers = producers_df.groupby('producers_list')['ROI'].mean().sort_values(ascending=False).head(3)

# Top Actor
all_actors_df = df.explode('actors_list')
top_actor = all_actors_df['actors_list'].value_counts().idxmax()
avg_profit = all_actors_df[all_actors_df['actors_list'] == top_actor]['Profit'].mean()

# --- THE FINAL REPORT PRINT ---
print("="*50)
print("       MOVIE PRODUCTION RECOMMENDATION REPORT        ")
print("="*50)
print(f"🏆 HIGHEST PROFIT MOVIE: {top_movie['title']}")
print(f"   Director: {top_movie['director']}")
print(f"   Net Profit: ${top_movie['Profit']:,.2f}")
print(f"\n💡 BEST INVESTMENT LANGUAGE: {language_roi.index[0]}")
print(f"   Average ROI: {language_roi.iloc[0]:.2f}x return")
print(f"\n🌟 MOST FREQUENT ACTOR: {top_actor}")
print(f"   Average Profit per Film: ${avg_profit:,.2f}")
print("="*50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# 1. Prepare the Genre Data
# We need to turn the 'genres' column into a clean list of names
def clean_genres(genre_str):
    try:
        data = ast.literal_eval(genre_str)
        return [item['name'] for item in data]
    except:
        return []

df['genres_list'] = df['genres'].apply(clean_genres)

# 2. "Explode" the genres so every genre (Action, Comedy, etc.) gets its own row
exploded_genres = df.explode('genres_list')

# 3. Calculate Average Profit per Genre
genre_profit = exploded_genres.groupby('genres_list')['Profit'].mean().sort_values(ascending=False)

# 4. Create the Bar Chart
plt.figure(figsize=(12, 8))
sns.barplot(x=genre_profit.values, y=genre_profit.index, palette="magma")

# 5. Add Labels and Title
plt.title('Which Genres Make the Most Money? (Average Profit)', fontsize=16)
plt.xlabel('Average Profit in Dollars ($)', fontsize=12)
plt.ylabel('Movie Genre', fontsize=12)

plt.show()
```
